```python
import pandas as pd

logs_handler_acl_java = "logs_handler_acl_java.txt"
logs_handler_acl_javascript = "logs_handler_acl_javascript.txt"
logs_handler_acl_go = "logs_handler_acl_go.txt"

logs_handler_log_java = "logs_handler_log_java.txt"
logs_handler_log_javascript = "logs_handler_log_javascript.txt"
logs_handler_log_go = "logs_handler_log_go.txt"

acl_java = pd.read_csv (logs_handler_acl_java, names = ["message"])
acl_js = pd.read_csv (logs_handler_acl_javascript, names = ["message"], sep = ";")
acl_go = pd.read_csv (logs_handler_acl_go, names = ["message"], sep = ";")
log_java = pd.read_csv (logs_handler_log_java, names = ["message"])
log_js = pd.read_csv (logs_handler_log_javascript, names = ["message"], sep = ";")
log_go = pd.read_csv (logs_handler_log_go, names = ["message"], sep = ";")
acl_js.head()
```

```python
import numpy as np
import re


#acl java 
list1 = []
list1 = acl_java["message"].tolist()
listMDB = []
for x in list1:
    if x.count("Read")>0:
        listMDB.append(x.split(" "))
listMDB2 = []
for x in listMDB:
    listMDB2.append(int(x[-2]))
acl_java_mdb = np.array(listMDB2)

list1 = []
list1 = acl_java["message"].tolist()
list2 = []
for x in list1:
    if x.count("example")>0:
        list2.append(x.split(" "))
list3 = []
for x in list2:
    match = re.match(r"([\d.]+)([a-zA-Z]+)", x[-1])
    list3.append(float(match.group(1)))
for x in range(len(list3)):
    if list3[x] < 10:
        list3[x] = int(list3[x] * 1000)
    else:
        list3[x] = int(list3[x])
acl_java_tot = np.array(list3)

list1 = []
list1 = acl_java["message"].tolist()
list2 = []
for x in list1:
    if x.count("access control")>0:
        list2.append(x.split(" "))
list3 = []
for x in list2:
    list3.append(int(x[-2]))
acl_java_fun = np.array(list3)




#acl js
list1 = []
list1 = acl_js["message"].tolist()
listMDB = []
for x in list1:
    if x.count("Read")>0:
        listMDB.append(x.split(" "))
listMDB2 = []
for x in listMDB:
    listMDB2.append(int(x[6]))
acl_js_mdb = np.array(listMDB2)


list1 = []
list1 = acl_js["message"].tolist()
list2 = []
for x in list1:
    if x.count("example")>0:
        list2.append(x.split(" "))
list3 = []
for x in list2:
    match = re.match(r"([\d.]+)([a-zA-Z]+)", x[-1])
    list3.append(float(match.group(1)))
for x in range(len(list3)):
    if list3[x] < 10:
        list3[x] = int(list3[x] * 1000)
    else:
        list3[x] = int(list3[x])
acl_js_tot = np.array(list3)


list1 = []
list1 = acl_js["message"].tolist()
list2 = []
for x in list1:
    if x.count("access control")>0:
        list2.append(x.split(" "))
list3 = []
for x in list2:
    list3.append(int(x[5]))
acl_js_fun = np.array(list3)




#acl go
list1 = []
list1 = acl_go["message"].tolist()
listMDB = []
for x in list1:
    if x.count("Read")>0:
        listMDB.append(x.split(" "))
listMDB2 = []
for x in listMDB:
    listMDB2.append(int(x[-2]))
acl_go_mdb = np.array(listMDB2)


list1 = []
list1 = acl_go["message"].tolist()
list2 = []
for x in list1:
    if x.count("example")>0:
        list2.append(x.split(" "))
list3 = []
for x in list2:
    match = re.match(r"([\d.]+)([a-zA-Z]+)", x[-1])
    list3.append(float(match.group(1)))
for x in range(len(list3)):
    if list3[x] < 10:
        list3[x] = int(list3[x] * 1000)
    else:
        list3[x] = int(list3[x])
acl_go_tot = np.array(list3)


list1 = []
list1 = acl_go["message"].tolist()
list2 = []
for x in list1:
    if x.count("access control")>0:
        list2.append(x.split(" "))
list3 = []
for x in list2:
    list3.append(int(x[5]))
acl_go_fun = np.array(list3)
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize = (20,8)) 
ax.set_title("Duration of service retrieval from MongoDB", fontsize = 40)
ax.set_xlabel('Invocations', fontsize = 34)
ax.set_ylabel('Duration [ms]', fontsize = 34)
ax.plot(acl_js_mdb, label = 'js function')
ax.plot(acl_java_mdb, label = 'java function')
ax.plot(acl_go_mdb, label = 'go function')
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
ax.grid(True, linewidth=0.5)
#ax.set_yticks(range(0,2300,100))
plt.yscale('log')
plt.legend(loc = 'upper right')
plt.rc('legend', fontsize = 25)
plt.show() 
```

```python
fig, ax = plt.subplots(figsize = (20,8)) 
ax.set_title("Example function total duration", fontsize = 40)
ax.set_xlabel('Invocations', fontsize = 34)
ax.set_ylabel('Duration [ms]', fontsize = 34)
ax.plot(acl_js_tot, label = 'js function')
ax.plot(acl_java_tot, label = 'java function')
ax.plot(acl_go_tot, label = 'go function')
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
ax.grid(True, linewidth=0.5)
ax.set_yticks(range(0,1200,100))
plt.yscale('log')
plt.legend(loc = 'upper right')
plt.rc('legend', fontsize = 25)
plt.show() 
```

```python
fig, ax = plt.subplots(figsize = (20,8)) 
ax.set_title("Access control function binding duration", fontsize = 40)
ax.set_xlabel('Invocations', fontsize = 34)
ax.set_ylabel('Duration [ms]', fontsize = 34)
ax.plot(acl_js_fun, label = 'js function')
ax.plot(acl_java_fun, label = 'java function')
ax.plot(acl_go_fun, label = 'go function')
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
ax.grid(True, linewidth=0.5)
#ax.set_yticks(range(0,1200,100))
#plt.yscale('log')
plt.legend(loc = 'upper right')
plt.rc('legend', fontsize = 25)
plt.show() 
```
